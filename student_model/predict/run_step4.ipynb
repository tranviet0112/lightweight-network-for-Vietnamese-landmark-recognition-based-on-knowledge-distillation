```
import torch
import torchvision
import torchvision.transforms as transforms
import glob
from predict import predict
import time
```

```
model_distil_CNN = '/content/drive/My Drive/Colab Notebooks/Resnet18-MODELC/fine_tuning_modelC/5CNN_ver2/step4_predict/best_C_5conv_ver2.pth.tar'
labels_url = '/content/drive/My Drive/Colab Notebooks/Resnet18-MODELC/fine_tuning_modelC/5CNN_ver2/step4_predict/labels_C_5conv_ver2.txt'
model_FNN =  '/content/drive/My Drive/Colab Notebooks/Resnet18-MODELC/fine_tuning_modelC/5CNN_ver2/step4_predict/model_best_C_5conv_ver2.pt'
correct = 0
total = 1500
inside = 0
total_time = 0
link_test_folder = '/content/drive/My Drive/Colab Notebooks/7CNN/final_dataset/test'
all_folders = list(sorted(glob.glob(f'{link_test_folder}/*')))
for folder in all_folders:
  t1 = time.time()
  inside += 1
  incorrect = 0
  label = folder.split('/')[-1]
  urls_img = list(sorted(glob.glob(f'{folder}/*.png')))
  for url in urls_img:
    predicted = predict(url, model_distil_CNN, labels_url, model_FNN)
    if predicted == label:
      correct += 1
      incorrect += 1
  t2 = time.time()
  intime = t2 - t1
  total_time += intime
  print('{}/30 complete {}, acc: {}, total time: {}'.format(inside, label, incorrect/50, intime))
print('Accuracy of the network on the testset images: {} , total time {}'.format(100 * correct / total, total_time))
```
